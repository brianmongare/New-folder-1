```python
import pandas as pd
import numpy as np
from pathlib import Path
import re
from sacremoses import MosesTokenizer
from typing import List, Tuple

class ParallelCorpusPreprocessor:
    def __init__(self, src_lang='en', tgt_lang='sw'):
        self.src_lang = src_lang
        self.tgt_lang = tgt_lang
        self.src_tokenizer = MosesTokenizer(lang=src_lang)
        self.tgt_tokenizer = MosesTokenizer(lang=tgt_lang)

    def clean_text(self, text: str) -> str:
        """Basic text cleaning."""
        text = text.strip()
        text = ' '.join(text.split())  # Normalize whitespace
        text = re.sub(r'[^\w\s\.,!?\-\'\":]', '', text)  # Keep basic punctuation
        return text

    def process_files(self, src_file: str, tgt_file: str, output_dir: str):
        # Create output directory
        Path(output_dir).mkdir(parents=True, exist_ok=True)

        # Read files
        with open(src_file, 'r', encoding='utf-8') as sf, \
             open(tgt_file, 'r', encoding='utf-8') as tf:
            src_lines = [line.strip() for line in sf.readlines()]
            tgt_lines = [line.strip() for line in tf.readlines()]

        # Create DataFrame
        df = pd.DataFrame({
            'source': src_lines,
            'target': tgt_lines
        })

        # Clean text
        print("Cleaning text...")
        df['source'] = df['source'].apply(self.clean_text)
        df['target'] = df['target'].apply(self.clean_text)

        # Remove empty lines and duplicates
        df = df.dropna()
        df = df.drop_duplicates()

        # Split into train/val/test (80/10/10)
        train_size = int(0.8 * len(df))
        val_size = int(0.1 * len(df))

        train_df = df[:train_size]
        val_df = df[train_size:train_size + val_size]
        test_df = df[train_size + val_size:]

        # Save processed files
        print("Saving processed files...")
        for name, data in [('train', train_df), ('val', val_df), ('test', test_df)]:
            data['source'].to_csv(f"{output_dir}/{name}.{self.src_lang}",
                                index=False, header=False)
            data['target'].to_csv(f"{output_dir}/{name}.{self.tgt_lang}",
                                index=False, header=False)

        print(f"Total pairs: {len(df)}")
        print(f"Training pairs: {len(train_df)}")
        print(f"Validation pairs: {len(val_df)}")
        print(f"Test pairs: {len(test_df)}")

# Initialize and run the preprocessor
preprocessor = ParallelCorpusPreprocessor()
preprocessor.process_files(
    src_file='/content/drive/MyDrive/en-sw.txt/OpenSubtitles.en-sw.en',
    tgt_file='/content/drive/MyDrive/en-sw.txt/OpenSubtitles.en-sw.sw',
    output_dir='processed_data'
)
```

```python
# Example translations
def translate_text(text, model, tokenizer):
    model.eval()
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model.generate(inputs["input_ids"], max_length=128, num_beams=4, early_stopping=True)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text

# Test with a few English sentences
test_sentences = [
    "Hello, how are you?",
    "I love programming.",
    "Where is the nearest hospital?"
]

print("Translations:")
for sentence in test_sentences:
    translation = translate_text(sentence, model, tokenizer)
    print(f"English: {sentence}")
    print(f"Swahili: {translation}")
    print()
```

```python
from sklearn.model_selection import train_test_split
import evaluate

# Ensure dataset is properly defined (replace with your data)
dataset = [
    ("Hello", "Habari"),
    ("Good morning", "Habari ya asubuhi"),
    ("Thank you", "Asante"),
    ("Goodbye", "Kwaheri"),
    ("Yes", "Ndio"),
    ("No", "Hapana")
]

# Split dataset into training and validation (80% train, 20% validation)
english_texts = [pair[0] for pair in dataset]
swahili_texts = [pair[1] for pair in dataset]

train_texts, val_texts, train_labels, val_labels = train_test_split(
    english_texts, swahili_texts, test_size=0.2, random_state=42
)

# Load BLEU metric
bleu = evaluate.load("sacrebleu")

# Generate translations for the validation set
predictions = []
references = []

for i in range(len(val_texts)):
    inputs = tokenizer(val_texts[i], return_tensors="pt", padding=True, truncation=True, max_length=128)
    outputs = model.generate(**inputs)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    predictions.append(translated_text)
    references.append([val_labels[i]])  # BLEU expects references as lists of lists

# Calculate BLEU score
results = bleu.compute(predictions=predictions, references=references)
print("BLEU Score:", results["score"])
```
